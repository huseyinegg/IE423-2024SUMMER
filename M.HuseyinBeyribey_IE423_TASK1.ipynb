# Now I will fit the regressin model

```python
#Which Columns exactly do we have?
dfFootball.columns.tolist()
```

```python
features = [
    'height', 'age', 'appearance', 'goals', 'assists', 'yellow cards',
    'red cards', 'goals conceded', 'clean sheets', 'minutes played',
    'days_injured', 'games_injured', 'award', 'highest_value',
    'position_encoded', 'winger'
]

X = dfFootball[features]
Y = dfFootball['current_value']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model = LinearRegression()

model.fit(X_train, Y_train)

intercept = model.intercept_
coefficients = model.coef_

(intercept, coefficients)
```

The array values represent the coefficients corresponding to the features in their specified order.

For instance, the coefficient for the variable 'Goals Conceded' is -313,269.97, for 'Red Cards' is -1,032,808.76, and for 'Goals' is 779,718.20. These values fit with the importance of each feature.

**Now I will find some statistical values to test success of fit.** I have used sckit_learn train_test_split method therefore I can easily test the regression model.

[train_test_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

Y_pred = model.predict(X_test)
r2 = r2_score(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)

r2, mse
```

r_2 indicates about 75% of the variability in the current value can be explained by the model which is good but not the best.

[r_2 score](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html)

Now I will test the model with arbitrary data I will create. Order of the data as in **features** variable.

```python
new_player_features = X_test.iloc[1]

predicted_current_value = model.predict([X_test.iloc[1]])

Acutal_value = Y_test.iloc[1]
predicted_current_value[0]
print("Acutal Value {}".format(Acutal_value))
print("Predicted Value {}".format(predicted_current_value[0]))
```

**Lastly, I will test the significance of model by F-test**


```python
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)
X_test_sm = sm.add_constant(X_test)

model_sm = sm.OLS(Y_train, X_train_sm).fit()

model_summary = model_sm.summary()
model_summary
```
