# Sentiment-analysis

````python
from transformers import pipeline

# Initialize the sentiment-analysis pipeline
mdlSnt_first = pipeline('sentiment-analysis')

# List of statements to analyze
prmStatement = [
    'It is nice to have you here in the US. I heard the weather is terrible in the UK.',
    'I am struggling with stress'
]

# Predict sentiment
results = mdlSnt_first(prmStatement)

# Display results
for statement, result in zip(prmStatement, results):
    label = result['label']
    score = result['score']
    print(f"Statement: \"{statement}\"")
    print(f"Sentiment: {label} (Score: {score:.4f})\n")
````

Words has Positive or Negative statements. Here probably
The word "nice" in "It is nice to have you here" contributes positively.
The word "terrible" in "I heard the weather is terrible in the UK" contributes negatively.

Now I will try with more statements.

````python
from transformers import pipeline

mdlSnt_english = pipeline('sentiment-analysis')

prmStatement_english = [
    'It is nice to have you here in the US. I heard the weather is terrible in the UK.',
    'I am struggling with stress.',
    'The presentation was fantastic, and the audience seemed very engaged.',
    'I feel overwhelmed by the amount of work I have to do.',
    'What a beautiful day! The sun is shining, and the sky is clear.',
    'I am disappointed with the service; it was a terrible experience.',
    'I love spending time with my family; they make me very happy.',
    'The food was delicious, but the service was slow.',
    'I can’t believe how rude the customer service was.',
    'I’m so excited about the upcoming vacation; it’s going to be amazing!'
]

# Predict sentiment
results = mdlSnt_english(prmStatement_english)

# Display results
for statement, result in zip(prmStatement_english, results):
    label = result['label']
    score = result['score']
    print(f"Statement: \"{statement}\"")
    print(f"Sentiment: {label} (Score: {score:.4f})\n")

# Determine statements with highest positive and negative scores
most_positive = max(results, key=lambda x: x['score'] if x['label'] == 'POSITIVE' else -1)
most_negative = max(results, key=lambda x: x['score'] if x['label'] == 'NEGATIVE' else -1)

# Print the most positive and most negative statements
print("Most Positive Statement:")
print(f"Statement: \"{prmStatement_english[results.index(most_positive)]}\"")
print(f"Sentiment: {most_positive['label']} (Score: {most_positive['score']:.4f})\n")

print("Most Negative Statement:")
print(f"Statement: \"{prmStatement_english[results.index(most_negative)]}\"")
print(f"Sentiment: {most_negative['label']} (Score: {most_negative['score']:.4f})")
````

Expectedly, word love, happy increases Positive score. On the other hand words dissapointed, terrible give NEGATIVE sentiment score.

````python
from transformers import pipeline

# Initialize question-answering pipelines for each model
models = {
    "BERT": "google-bert/bert-large-uncased-whole-word-masking-finetuned-squad",
    "RoBERTa": "deepset/roberta-base-squad2",
    "DistilBERT": "distilbert-base-uncased-distilled-squad",
    "ALBERT": "ktrapeznikov/albert-xlarge-v2-squad-v2",
    "T5": "t5-large",
    "ELECTRA": "google/electra-large-discriminator"
}


questions = [
    "What is the main topic of the conference in 1964, at a conference on Nietzsche in Royaumont ? ",
    "who took care of nietzsche's archives near the death of the nietzsche and in the near future of his death?",
    "How did Elisabeth Förster-Nietzsche's handling of Friedrich Nietzsche's works alter his philosophical reputation?",
    "What are the potential consequences of a philosopher's unpublished works being included or excluded from their archive?",
    "What challenges do scholars face when dealing with philosophical archives?",
    "In what ways can the creation and management of an archive be seen as a political act?"
]

answers = {}

for model_name, model_path in models.items():
    qa_pipeline = pipeline("question-answering", model=model_path, tokenizer=model_path)
    model_answers = []
    for question in questions:
        result = qa_pipeline(question=question, context=context)
        model_answers.append(result['answer'])
    answers[model_name] = model_answers

answers
````

RoBERTa model works the best it answer first three questions true. ALBERT also was able to answer second question right. As a result, question-answering models only works well whenever the answer is short at most 5-6 words.

````python
from google.colab import drive
drive.mount('/content/drive')
import json
import torch
import transformers
from transformers import (AutoTokenizer,
                          AutoModelForCausalLM,
                          BitsAndBytesConfig,
                          pipeline)

#HF account Configuration
config_data = json.load(open("/content/drive/MyDrive/Colab Notebooks/data/config.json"))
HF_TOKEN = config_data["HF_TOKEN"]
````

**Quantisation Configuration**

````python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
````

#Meta-Llama-3.1-8B-Instruct

https://huggingface.co/meta-llama/Meta-Llama-3.1-8B-Instruct

First Let's try small 8B model.

**Loading the Tokenizer and the LLM**


````python
model_name = "meta-llama/Meta-Llama-3.1-8B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name,
                                          token=HF_TOKEN)

tokenizer.pad_token = tokenizer.eos_token
````

````python
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    quantization_config=bnb_config,
    token=HF_TOKEN
)
````

````python
text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=128
)
def generate_responses(questions, line_length=80):
    responses = []
    for question in questions:
        prompt = question
        sequences = text_generator(prompt)
        gen_text = sequences[0]["generated_text"]
        # Adding line breaks after every 'line_length' characters
        wrapped_text = "\n".join(gen_text[i:i+line_length] for i in range(0, len(gen_text), line_length))
        responses.append(wrapped_text)
    return "\n\n".join(responses)

questions = [
    "What is the main topic of the conference in 1964, at a conference on Nietzsche in Royaumont ? ",
    "who took care of nietzsche's archives near the death of the nietzsche and in the near future of his death?",
    "How did Elisabeth Förster-Nietzsche's handling of Friedrich Nietzsche's works alter his philosophical reputation?",
    "What are the potential consequences of a philosopher's unpublished works being included or excluded from their archive?",
    "What challenges do scholars face when dealing with philosophical archives?",
    "In what ways can the creation and management of an archive be seen as a political act?"
]
compiled_response_llama8b = generate_responses(questions)
print(compiled_response_llama8b)
````
